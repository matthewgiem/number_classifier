```python
%matplotlib inline
import numpy as np
import scipy
import pandas as pd
import sklearn
import seaborn
import matplotlib
import matplotlib.pyplot as plt
import librosa
import librosa.display
import IPython.display as ipd
from glob import glob
```

```python
data_dir = './data/new_numbers'
audio_files = glob(data_dir + '/*.wav')
```

```python
audio, sfreq = librosa.load(audio_files[100])
time = np.arange(0, len(audio))/sfreq
```

```python
fig, ax = plt.subplots()
ax.plot(time, audio)
ax.set(xlabel = 'Time (s)', ylabel = 'Sound Amplitude')
plt.show()
```

```python
%%time

for file in range(0, len(audio_files), 1):
    audio, sfreq = librosa.load(audio_files[file])
    time = np.arange(0, len(audio))/sfreq
    
    fig, ax = plt.subplots()
    ax.plot(time, audio)
    ax.set(xlabel = 'Time (s)', ylabel = 'Sound Amplitude')
    plt.show()
```
